```python
import pandas as pd
data=pd.read_csv('OnlineRetail.csv',encoding="ISO-8859-1")
data
```

```python
#Đọc bộ dữ liệu, cho biết số dòng, số cột và kiểu dữ liệu của các thuộc tính.
print("Số dòng là",data.shape[0])
print("Số cột là", data.shape[1])
data.info()
data.rename(columns={'Unnamed: 0':'Invoice'},inplace=True)
display(data)
```

```python
#Trích xuất dữ liệu các cột Description và Quantity lưu vào file OnlineRetail.csv
data1= data.loc[:,['Description','Quantity']]
display(data1)
data1.to_csv('demo_OnlineRetail.csv')
```

```python
#Trích xuất dữ liệu 1000 dòng đầu tiên lưu vào file OnlineRetail.xlsx
data2=data.iloc[1:1001]
display(data2)
data2.to_excel('demo_OnlineRetail.xlsx')
```

```python
#Trích xuất các bản ghi có số lượng từ 10 trở lên lưu vào file OnlineRetail.h5
data3=data.loc[data.Quantity>10]
display(data3)
data3.to_hdf('OnlineRetail.h5', 'table')
```

```python
#Trích xuất dữ liệu các phần tử từ dòng 1000 đến dòng 2000, các cột Quantity, InvoiceDate, UnitPrice lưu vào file OnlineRetail.json
data4=data.loc[1001:2001,['Description','Quantity',]]
display(data4)
data4.to_json('demo_OnlineRetail.json',orient='columns')
```

```python
#Trích xuất các bản ghi có số hóa đơn ‘536365’ lưu vào file OnlineRetail.html
data5=data.loc[data.Invoice == 536365]
display(data5)
data5.to_html('demo_OnlineRetail.html.html')
```
