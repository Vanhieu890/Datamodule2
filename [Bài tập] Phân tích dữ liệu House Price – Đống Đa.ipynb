```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn import preprocessing as pp
data=pd.read_excel('house_price_dống-da.xlsx')
data
```

```python
#Đọc bộ dữ liệu, cho biết số dòng, số cột và kiểu dữ liệu của các thuộc tính.
data.info()
print('So dong du lieu la',data.shape[0])
print('So cot du lieu la',data.shape[1])
```

```python
data.columns
```

```python
#Phát hiện các cột chứa dữ liệu khiếm khuyết
print('Các cột chứa dữ liệu khuyết thiếu là:')
for index, column in enumerate(data.columns):
    temp=data.loc[:,column].isnull()
    if temp.sum()>0:
        print('Cột',data.columns[index],'chứa', temp.sum(), 'giá trị khuyết thiếu')
        
```

```python
#Lọc ra các bản ghi bán nhà riêng tại phường Trung liệt hoặc phường Khâm Thiên
#loc du lieu khiem khuyet cua type_of_land
data.loc[:,'type_of_land'].unique()
```

```python
data1=data[(data['type_of_land']=='Bán nhà riêng\n ') | (data['type_of_land']=='Bán nhà riêng')]
data1
data2=data1[(data1['ward_name']=='Phường Trung Liệt')|(data1['ward_name']=='Phường Khâm Thiên')]
data2
```

```python
#Lọc các thông tin Địa chỉ, Giá, Hướng nhà, Hướng ban công của các bản ghi có giấy chứng nhận sổ đỏ và có 3 phòng ngủ trở lên.
data4 = data[['address', 'price','house_direction','balcony_direction', 'balcony_direction','bedroom']]
data4
```

```python
data4.loc[:,'bedroom'].unique()
```

```python
data4.dropna(subset=['bedroom'],inplace=True)
data4
```

```python
#Với mỗi loại nhà đất, tính trung bình cộng giá cũng như giá lớn nhất và giá nhỏ nhất.
data.head(10)
data.loc[:,'type_of_land'].unique()
```

```python
data5=data[['price','type_of_land']]
data5['type_of_land'].replace('Bán nhà riêng\n ','Bán nhà riêng',inplace=True)
data5['type_of_land'].replace('Bán nhà mặt phố\n ','Bán nhà mặt phố',inplace=True)
data5['type_of_land'].replace('Bất động sản khác\n ','Bất động sản khác',inplace=True)
data5
```
