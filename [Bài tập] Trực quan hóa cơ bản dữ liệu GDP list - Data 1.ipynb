```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime
data=pd.read_csv('GDPlist.csv', encoding='ISO-8859-1')
data
```

```python
data1=data
tg=data1.loc[data1.Continent=='South America']
tg
```

```python
y=tg.index.get_level_values(0)
plt.bar(tg['Country'],y)
plt.title('GDP of the continent in South America')
plt.xlabel('Country')
plt.ylabel('GDP')
plt.show()
```
