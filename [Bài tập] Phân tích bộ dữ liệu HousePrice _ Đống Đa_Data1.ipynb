```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
print('Khai báo thư viện thành công')
```

```python
data=pd.read_csv('house_price_Dống-Da_Hà-Nội_subdata.csv')
data
```

```python
data.info()
```

```python
plt.scatter(data['area'],data['price'])
plt.title('Relationship between area and price',fontsize=16,color='r')
plt.xlabel('area',fontsize=14)
plt.ylabel('price',fontsize=14)
plt.show()
```

```python
plt.scatter(data['bedroom'],data['price'])
plt.title('Relationship between bedroom and price',fontsize=16,color='r')
plt.xlabel('bedroom',fontsize=14)
plt.ylabel('price',fontsize=14)
plt.show()
```

```python
plt.scatter(data['toilet'],data['price'])
plt.title('Relationship between toilet and price',fontsize=16,color='r')
plt.xlabel('toilet',fontsize=14)
plt.ylabel('price',fontsize=14)
plt.show()
```
