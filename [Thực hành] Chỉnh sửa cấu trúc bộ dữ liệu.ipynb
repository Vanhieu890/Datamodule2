```python
import pandas as pd
data=pd.read_csv('FoodPrice_in_Turkey.csv', encoding="ISO-8859-1")
display(data)
```

```python
#2. Đổi tên cột thuộc tính
```

```python
data.rename(columns={'Place':'Địa chỉ','ProductId':'Mã sản phẩm','ProductName':'Tên sản phẩm','UmId':'Số lượng','UmName':'Đơn vị','Month':'Tháng','Year':'Năm','Price':'Giá'},inplace=True)
data
```

```python
#3. Thêm một cột mới trong DataFrame
```

```python
# Thêm một cột mới với tất cả các giá trị rỗng NaN
data['new_column']='NaN'
data
```

```python
# Thêm cột giảm giá 10% cho tất cả các bản ghi
data['Giảm giá']='10%'
data
```

```python
# Thêm cột giảm giá 12% cho tất cả các bản ghi
data.insert(10,'GIảm giá 2',pd.Series('12%',index=data.index))
data
```

```python
#4. Thêm một dòng mới vào trong DataFrame
data=data.append({'Địa chỉ':'Hà Nội','Mã sản phẩm':'RR','Tên sản phẩm':'Rice','Số lượng':10,'Đơn vị':'KG','Tháng':6,'Năm':2021,'Giá':84.3785,'new_comlumn':'NaN','Giảm giá':'10%','Giảm giá 2':'12%'},ignore_index=True)
data
```

```python
#5. Xóa một cột trong DataFrame
data1=data
del data1['new_column']
data1
```
