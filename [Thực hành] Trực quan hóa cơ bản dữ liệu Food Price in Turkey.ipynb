```python
#Vẽ biểu đồ cột so sánh giá gạo (Rice-Retail) tháng 12 năm 2019 của Ankara, Istanbul, Izmir và National Average.
#Vẽ biểu đồ đường phân tích xu hướng giá gạo (Rice-Retail) trung bình cả nước (National Average) trong năm 2019 tại Thổ Nhĩ Kì.
#Vẽ biểu đồ Scatter phân tích mối liên quan giữa giá gạo và giá gas trung bình quốc gia (National Average) tại Thổ Nhĩ Kì.
```

```python
import pandas as pd 
import matplotlib.pyplot as plt
data=pd.read_csv('FoodPrice_in_Turkey.csv')
data
```

```python
#Vẽ biểu đồ cột so sánh giá gạo (Rice-Retail) tháng 12 năm 2019 của Ankara, Istanbul, Izmir và National Average.
data1=data[(data['Year']==2019)& (data['Month']==12)&(data['ProductName']=='Rice - Retail')]
data1
```

```python
plt.bar(data1['Place'],data1['Price'])
plt.show()
```

```python
plt.bar(data1['Place'],data1['Price'],width=0.5)#width: thay đổi chiều rộng các cột
plt.title('Rice Price in 12/2019', fontsize=16,color='r')
plt.xlabel('Place', fontsize=14)
plt.ylabel('Price',fontsize=14)
plt.show()
print('Giá gạo Tháng 12/2019 của Istanbul vầ Izmir tương đương và cao hơn giá gạo trung bình quốc gia và Ankara')
print('Giá gạo tháng 12/2019 tại Ankara thấp nhất')
```

```python
#Vẽ biểu đồ đường phân tích xu hướng giá gạo (Rice-Retail) trung bình cả nước (National Average) 
#trong năm 2019 tại Thổ Nhĩ Kì.
data2=data[(data['Year']==2019)& (data['Place']=='National Average')&(data['ProductName']=='Rice - Retail')]
data2
```

```python
plt.plot(data2['Price'], data2['Month'])
plt.show()
```

```python
plt.plot(data2['Price'],data2['Month'],linewidth=2,marker='*',markersize=10,markerfacecolor='red',markeredgecolor='blue',markeredgewidth=2)
plt.title('Rice Price of Nation Average in 2019',fontsize=16, color='r')
plt.xlabel('Price', fontsize=14)
plt.ylabel('Month',fontsize=14)
plt.show()
print('Giá gạo trung bình quốc gia năm 2019 tăng liên tục từ tháng 1 cho tới tháng 9, tháng 10, 11 có giảm nhẹ. Tuy nhiên, giá gạo lại tiếp tục tăng trở lại vào tháng 12/2019')
print('Giá gạo trung bình quốc gia nhìn chung có xu hướng tăng và dự kiến vẫn có thể tiếp tục tăng vào năm 2020.')
```

```python
#Vẽ biểu đồ Scatter phân tích mối liên quan giữa giá gạo và giá gas trung bình quốc gia (National Average) tại Thổ Nhĩ Kì.
data3=data[(data['Place']=='National Average')&(data['ProductName']=='Rice - Retail')&(data['Year']==2019)]
data4=data[(data['Place']=='National Average')&(data['ProductName']=='Fuel (gas) - Retail')&(data['Year']==2019)]
display(data3)
display(data4)
                                                                                            
```

```python
plt.scatter(data3['Price'],data4['Price'])
plt.show()
```

```python
plt.scatter(data3['Price'],data4['Price'])
plt.title('Relationship between Rice and Gas of National Average',fontsize=16,color='r')
plt.xlabel('Rice',fontsize=14)
plt.ylabel('Gas',fontsize=14)
plt.show()
print('Giá Gas có ảnh hưởng khá mạnh tới giá Gạo')
print('Giá Gạo tăng rất nhanh khi giá Gas tăng lên')
print('Xu hướng tăng có thể coi là tuyến tính')
print('Trong dữ liệu có 1 điểm bất thường: giá gạo thấp nhưng giá Gas lại rất cao. Điều này xảy ra có thể do 1 sự kiện bất thường hoặc do nhập liệu.')
```
