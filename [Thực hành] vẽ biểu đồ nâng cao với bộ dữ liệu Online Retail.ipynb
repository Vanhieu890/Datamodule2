```python
import pandas as pd
import seaborn as sns
data=pd.read_csv('2.OnlineRetail.csv',encoding='ISO-8859-1')
data
```

```python
data.info()
```

```python
data=data.dropna()
```

```python
data['Revenue']=data['Quantity']*data['UnitPrice']
data
```

# Vẽ biểu đồ phân bố

```python
#Vẽ biểu đồ phân bố cho giá sản phẩm
sns.violinplot(y='Quantity',data=data);
```

```python
#biểu đồ phân bố cho tổng giá mỗi sản phẩm ở các đơn
sns.violinplot(y='Revenue',data=data)
```

```python
#Tính số lượng sản phẩm ở mỗi đơn hàng
data2=data.groupby(['InvoiceNo'])['Quantity'].sum().reset_index()
data2
```

```python
#Biểu đồ phân bố số lượng sản phẩm trên mỗi đơn
sns.violinplot(y='Quantity',data=data2)
```

Vẽ biểu đồ tần số

```python
data3=data.groupby(['Country'])['Quantity'].sum().reset_index()
data3
```

```python
#Loại bổ hóa đơn trùng lặp

data4=data.drop_duplicates(subset='InvoiceNo')
```

```python
#Vẽ biểu đồ tần số cho số hóa đơn theo quốc gia
sns.countplot(x='Country',data=data4)
```

Ve biểu đồ box plot

```python
#Vẽ biểu đồ box plot cho thuộc tính giá sản phẩm
sns.boxplot(x=data['UnitPrice']);
```

```python
#Vẽ biểu đồ cho số lượng sản phẩm ở mỗi đơn
sns.boxplot(x=data2['Quantity'])
```
